```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(url)


print("Data Shape:", data.shape)

print("Missing values in data:\n", data.isnull().sum())

print("Data types in data:\n", data.dtypes)

print("Zeros in data:\n", (data == 0).sum())

selected_data = data.loc[:, ['Name', 'Age', 'Survived']] 

sorted_data = data.sort_values(by='Age')

print("Data description:\n", data.describe())

print("Unique values in data:\n", data.nunique())

data['Age'] = data['Age'].fillna(-1).astype('int')  

print(data.head())
```
